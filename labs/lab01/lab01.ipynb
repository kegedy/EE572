```python
# Part1_unknown3
#
def seriesR_L(R,L,Z0,Ei,t):
    Tao = L/(R+Z0)
    const = (R-Z0)/(R+Z0)
    x0 = t[t<=-0.00000005]
    x1 = t[(t>-0.00000005) & (t<=0)]
    x2 = t[t>0]
    y0 = lambda x: np.zeros(len(x))
    y1 = lambda x: Ei*np.ones(len(x))
    y2 = lambda x: Ei*((1+const)+(1-const)*np.exp(-x/Tao))
    return y0(x0).tolist()+\
            y1(x1).tolist()+\
            y2(x2).tolist()

fig,ax = plt.subplots(figsize=(8,4))
ax.plot(t,seriesR_L(R,L,Z0,Ei,t))
ax.set_title('seriesR_L')
ax.set_xlabel('t');
```

![](images/shuntR_C.png)

```python
def shuntR_C(R,C,Z0,Ei,t):
    Tao = C*(Z0*R)/(Z0+R)
    const = (R-Z0)/(R+Z0)
    x0 = t[t<=-0.00000005]
    x1 = t[(t>-0.00000005) & (t<=0)]
    x2 = t[t>0]
    y0 = lambda x: np.zeros(len(x))
    y1 = lambda x: Ei*np.ones(len(x))
    y2 = lambda x: Ei*((1+const)*(1-np.exp(-x/Tao)))
    return y0(x0).tolist()+\
            y1(x1).tolist()+\
            y2(x2).tolist()

fig,ax = plt.subplots(figsize=(8,4))
ax.plot(t,shuntR_C(R,C,Z0,Ei,t))
ax.set_title('shuntR_C')
ax.set_xlabel('t');
```

```python
# Part1_unknown1
#
def shuntR_L(R,L,Z0,Ei,t):
    Tao = L*(Z0+R)/(Z0*R)
    const = (R-Z0)/(R+Z0)
    x0 = t[t<=-0.00000005]
    x1 = t[(t>-0.00000005) & (t<=0)]
    x2 = t[t>0]
    y0 = lambda x: np.zeros(len(x))
    y1 = lambda x: Ei*np.ones(len(x))
    y2 = lambda x: Ei*((1+const)*np.exp(-x/Tao))
    return y0(x0).tolist()+\
            y1(x1).tolist()+\
            y2(x2).tolist()

fig,ax = plt.subplots(figsize=(8,4))
ax.plot(t,shuntR_L(R,L,Z0,Ei,t))
ax.set_title('shuntR_L')
ax.set_xlabel('t');
```

```python
# Part1_unknown2
#
def seriesR_C(R,C,Z0,Ei,t):
    Tao = (R+Z0)*C
    const = (R-Z0)/(R+Z0)
    x0 = t[t<=-0.00000005]
    x1 = t[(t>-0.00000005) & (t<=0)]
    x2 = t[t>0]
    y0 = lambda x: np.zeros(len(x))
    y1 = lambda x: Ei*np.ones(len(x))
    y2 = lambda x: Ei*(2-(1-const)*np.exp(-x/Tao))
    return y0(x0).tolist()+\
            y1(x1).tolist()+\
            y2(x2).tolist()

fig,ax = plt.subplots(figsize=(8,4))
ax.plot(t,seriesR_C(R,C,Z0,Ei,t))
ax.set_title('seriesR_C')
ax.set_xlabel('t');
```
